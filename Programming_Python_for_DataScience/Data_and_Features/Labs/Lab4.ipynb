```python
import pandas as pd
import numpy as np
```

```python
nhl = pd.read_csv('NHL_2014-15.csv', sep=',')
nhl.head(5)
```

Get rid of the 'RK' column:

```python
nhl = nhl.drop(columns = ['RK']) # or nhl.drop(0)
nhl = nhl.reset_index(drop=True)
nhl.head(5)
```

Check the data type of all columns, and ensure those that should be numeric are numeric.

```python
# fixing some strings: replacing the , to a '' and adding the seconds to the time at column TOI/G
nhl.SHFT = nhl['SHFT'].str.replace(',','')
nhl['TOI/G'] = nhl['TOI/G'] + ':00'
nhl.head(5)
```

```python
# fixing the data types:  converting objetc to numeric and time type
print(nhl.dtypes)

nhl.SHFT = pd.to_numeric(nhl.SHFT, errors='coerce')
nhl.FW = pd.to_numeric(nhl.FW, errors='coerce')
# Time On Ice Per Game (TOI/G)
nhl['TOI/G'] = pd.to_timedelta(nhl['TOI/G'], errors='coerce')

print(nhl.dtypes)
nhl.head(5)
```
