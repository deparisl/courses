For this assignment, you'll be using the seeds data set, generated by recording X-Ray measurements of various wheat kernels. Start by opening up the starter code located in /Module3/Module3 - Lab1.ipynb, and reading through it. Then, write code that...

1 - Loads the seeds dataset, located at /Module3/Datasets/wheat.data into a dataframe.

2 - Creates a slice of your dataframe that only includes the area and perimeter features.

3 - Creates another slice that only includes the groove and asymmetry features.

4 - Creates a histogram for the 'area and perimeter' slice, and another histogram for the 'groove and asymmetry' slice. Set the optional display parameter: alpha=0.75.


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
plt.style.use('ggplot')
```

```python
# Load up the wheat seeds dataset into a dataframe.

seeds = pd.read_csv('wheat.txt', sep=',')
seeds.head(5)
```

```python
# Create a slice from your dataframe and name the variable s1. It should only include the area and perimeter features.

s1 = { 'area': seeds.area, 'perimeter': seeds.perimeter}
s1 = pd.DataFrame(s1)
s1.fillna(0)
```

```python
# Create another slice of from dataframe called it s2 this time. Slice out only the groove and asymmetry features:

s2 = {'groove': seeds.groove, 'asymmetry': seeds.asymmetry}
s2  = pd.DataFrame(s2)
s2['groove'] = s2['groove'].astype('str').str.replace('nan','0')
s2['groove'] = pd.to_numeric(s2.groove, errors='coerce')
```

```python
# Create a histogram plot using the first slice, and another histogram plot using the second slice. Be sure to set alpha=0.75.
plt.hist(s1.area, alpha=0.75, bins=10, label='Area')
plt.hist(s1.perimeter, alpha=0.75, bins=10, label='Perimeter')
plt.xlabel('ID')
plt.ylabel('#')
plt.legend()

plt.show()
```

```python
plt.hist(s2.groove, alpha=0.75, label='Groove')
plt.hist(s2.asymmetry, alpha=0.75, label='Asymmetry')
plt.xlabel('ID')
plt.ylabel('#')
plt.legend()
plt.show()
```
