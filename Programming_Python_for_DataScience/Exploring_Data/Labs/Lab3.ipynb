```python
%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load up the wheat seeds dataset into a dataframe.
#matplotlib.style.use('ggplot')
seeds = pd.read_csv('wheat.txt', sep=',')
seeds.head(5)
```

```python
# Create a slice from your dataframe.

s1 = { 'area': seeds.area, 'perimeter': seeds.perimeter, 'asymmetry': seeds.asymmetry}
s1 = pd.DataFrame(s1)
s1 = s1.fillna(0)

s2 = { 'width': seeds.width, 'groove': seeds.groove, 'length': seeds.length}
s2 = pd.DataFrame(s2)
s2 = s2.fillna(0)
```

```python
# 3D plots

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Area')
ax.set_ylabel('Perimeter')
ax.set_zlabel('Asymmetry')

ax.scatter(s1.area, s1.perimeter, s1.asymmetry, c='r', marker='.')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Width')
ax.set_ylabel('Groove')
ax.set_zlabel('Length')

ax.scatter(s2.width, s2.groove, s2.length, c='r', marker='.')
plt.show()
```
