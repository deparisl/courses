```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
# load the seeds table and drop the id column.

seeds = pd.read_csv('wheat.txt', sep=',', usecols=[i for i in range(1,9)])
seeds.fillna(0).head(5)
```

```python
# compute the correlation matrix: types of methods: ‘pearson’, ‘kendall’, ‘spearman’ or callable
# DataFrame.corr(method='pearson', min_periods=1)

seeds.corr()
```

```python
plt.imshow(seeds.corr(), cmap=plt.cm.Blues, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(seeds.columns))]
plt.xticks(tick_marks, seeds.columns, rotation='vertical')
plt.yticks(tick_marks, seeds.columns)

plt.show()
```
